```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
#import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'/Users/brucebeck/Desktop/RDBMS_Work/ATA_PortFolioProject4/movies.csv')

# Now we need to limit the maximum rows to 20
pd.set_option('display.max_rows',20)
```

```python
# Taking a inital look at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Dropping rows with null values

df = df.dropna()
```

```python
# Verifying NULL values are gone

df[df.isnull().any(axis=1)]

# Yep, they are gone
```

```python
# Changing Data Types for Budget & Gross Columns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# Verifying Data Types for budget and gross columns have changed

print(df.dtypes)
```

```python
# Let's do another check for NULL values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
# NICE! This is what I wanted to see.  
# Hope I did't break anything further along in the script.
```

```python
# Now let's get rid of any duplicate rows

df.drop_duplicates()
```

```python
# Correlation matix graphic for Pearson method

correlation_matrix = df.corr(method='pearson',numeric_only ='true')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#Numerizing all data

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
# Generating a new correlation matrix bases on numerized data

correlation_matrix = df_numerized.corr(method='pearson',numeric_only ='true')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Taking another look at df_numerized correlation

df_numerized.corr()
```

```python
# Looking at correlation pairs - unsorted 

correlation_matrix = df_numerized.corr()

corr_pairs = correlation_matrix.unstack()

corr_pairs
```

```python
#Sorting correlation pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Defining high correlation of sorted pairs

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
